```python
import pandas as pd
import altair as alt

# Load data from Excel
df = pd.read_excel("/Users/praveenzirali/active_projects/Crcsim Runs/Scatter Plot Data.xlsx")

# Plotting
scatter = alt.Chart(df).mark_circle(size=60).encode(
    x='FQHC1_baseline_implementation_LYS:Q',
    y='FQHC4_baseline_implementation_LYS:Q',
    color=alt.condition(alt.datum['FQHC1_baseline_implementation_LYS'] > alt.datum['FQHC4_baseline_implementation_LYS'],
                        alt.value('red'), alt.value('green')),  # Color based on condition
    tooltip=['FQHC1_baseline_implementation_LYS:Q', 'FQHC4_baseline_implementation_LYS:Q']
).properties(
    width=600,
    height=400,
    title='Comparison of FQHC1 vs FQHC4: Life Years Saved'
)

# Add legend
legend = alt.Chart(pd.DataFrame({'color': ['red', 'green'], 'label': ['FQHC1', 'FQHC4']})).mark_point(size=100).encode(
    y=alt.Y('label:N', axis=alt.Axis(orient='right')),
    color=alt.Color('color:N', legend=None)
).properties(
    width=600,
    height=400
)

(scatter + legend).configure_legend(
    orient='top-right'
)
```

```python
import altair as alt
import pandas as pd

# Read the data from the CSV file
data = pd.read_csv("/Users/praveenzirali/active_projects/Book4.csv")

# Create the first scatter plot for FQHC4_baseline_implementation_LYS vs FQHC1_baseline_implementation_LYS
scatter_plot_1 = alt.Chart(data).mark_circle(opacity=0.7).encode(
    x='FQHC4_baseline_implementation_LYS:Q',
    y='FQHC1_baseline_implementation_LYS:Q',
    color=alt.Color('Source:N', scale=alt.Scale(domain=['FQHC1', 'FQHC4'], range=['red', 'green']),
                    legend=alt.Legend(title='Legend'))
).properties(
    title='FQHC4_baseline_implementation_LYS vs FQHC1_baseline_implementation_LYS'
)

# Create the second scatter plot for FQHC1_baseline_implementation_LYS vs FQHC4_baseline_implementation_LYS
scatter_plot_2 = alt.Chart(data).mark_circle(opacity=0.7).encode(
    x='FQHC1_baseline_implementation_LYS:Q',
    y='FQHC4_baseline_implementation_LYS:Q',
    color=alt.Color('Source:N', scale=alt.Scale(domain=['FQHC1', 'FQHC4'], range=['red', 'green']),
                    legend=alt.Legend(title='Legend'))
).properties(
    title='FQHC1_baseline_implementation_LYS vs FQHC4_baseline_implementation_LYS'
)

# Layer the scatter plots
layered_scatter_plot = alt.layer(scatter_plot_1, scatter_plot_2, data=data).resolve_scale(
    color='independent'
)

# Display the layered plot
layered_scatter_plot.show()
```
